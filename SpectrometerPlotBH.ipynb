```python
fname = "/home/julian/Downloads/Example UV file.txt.txt"
import pandas
k = pandas.read_csv(fname, skiprows=4, sep="\t")
k.columns = ["t"]+["c%d"%i  for i in range(0, len(k.columns)-1)]
k.t = pandas.to_datetime(k.t)
k.set_index("t", inplace=True)
k.index.name = None
k.head()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(15,11))
plt.subplot(2,1,1)
plt.plot(k.c550)
plt.plot(k.c1900)
plt.legend()
plt.subplot(2,1,2)

for i in range(0, len(k), 150):
    plt.plot(list(k.iloc[i]), label=str(k.index[i]))
plt.legend()
```

```python
plt.figure(figsize=(11,5))
plt.imshow(k, vmin=2000, vmax=4000)
```
