```python
import pandas

f = "hg1a1b.txt"
k = pandas.read_csv(f, sep=" ", names=["x","y","z","r","g","b","nx","ny","nz"])

```

```python
k.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
b = k.iloc[::1000]
plt.scatter(b.x,b.z)
```

```python
xrange = min(k.x), max(k.x)
yrange = min(k.y), max(k.y)
zrange = min(k.z), max(k.z)

xrect = [yrange[0],yrange[1],yrange[1],yrange[0],yrange[0]]
yrect = [zrange[0],zrange[0],zrange[1],zrange[1],zrange[0]]
```

```python
x = -16
dx=0.01
b = k[(k.x>x) & (k.x<x+dx)]
plt.figure(figsize=(11,11))
plt.plot(xrect,yrect,color="r")
plt.scatter(b.y,b.z,marker=".")
```

```python
import subprocess, os
import numpy
os.mkdir("imgs")
```

```python
for i, x in enumerate(numpy.arange(-20,20,0.1)):
    print(x)
    plt.figure(figsize=(11,11), frameon=False)
    dx=0.01
    b = k[(k.x>x) & (k.x<x+dx)]
    plt.plot(xrect,yrect,color="r")
    plt.scatter(b.y,b.z,marker=".")
    plt.title("xslice %d" % x)
    plt.savefig("imgs/slc_%d.png" % i, format='png', pad_inches=0.0, bbox_inches='tight')
    plt.close()

# then 
# ffmpeg -start_number 0 -i imgs/slc_%d.png -vcodec mpeg4 U2melt.avi
```
